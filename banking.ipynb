```python
import pandas  as pd 
```

```python
data = pd.read_csv("D:/ds/CSV_Files/banking.csv")
```

```python
data.columns
```

```python
data.info()
```

```python
data['job'].unique()
```

```python
data['loan'] = data['loan'].map({'yes': 1, 'no': 0})
```

```python
data['housing'].unique()
```

```python
data['housing'] = data['housing'].map({'yes': 1, 'no': 0})
data.head()
```

```python
data.dropna(subset=['housing'], inplace=True)
```

```python
data.dropna(subset=['loan'], inplace=True)
```

```python
data['poutcome'].unique()
```

```python
data['poutcome'] = data['poutcome'].map({'success': 1, 'nonexistent': 0, 'failure': -1})
data.head()
```

```python
data.drop(columns=['age', 'marital', 'education', 'contact', 'month', 'day_of_week', 'campaign', 'previous', 'pdays', ], inplace=True)
```

```python
data.columns
```

```python
job_data = pd.get_dummies(data['job'], drop_first=True)
```

```python
veri = {False: 'False', True: 'True'}
job_data['blue-collar'] = job_data['blue-collar'].map(veri)
job_data['blue-collar'] = job_data['blue-collar'].map({'False': 0, 'True': 1})

job_data['entrepreneur'] = job_data['entrepreneur'].map(veri)
job_data['entrepreneur']= job_data['entrepreneur'].map({'False': 0, 'True': 1})

job_data['housemaid']= job_data['housemaid'].map(veri)
job_data['housemaid']= job_data['housemaid'].map({'False': 0, 'True': 1})

job_data['management'] = job_data['management'].map(veri)
job_data['management'] = job_data['management'].map({'False': 0, 'True': 1})

job_data['retired'] = job_data['retired'].map(veri)
job_data['retired']= job_data['retired'].map({'False': 0, 'True': 1})

job_data['self-employed'] = job_data['self-employed'].map(veri)
job_data['self-employed'] = job_data['self-employed'].map({'False': 0, 'True': 1})

job_data['services'] = job_data['services'].map(veri)
job_data['services'] = job_data['services'].map({'False': 0, 'True': 1})

job_data['student'] = job_data['student'].map(veri)
job_data['student'] = job_data['student'].map({'False': 0, 'True': 1})

job_data['technician'] = job_data['technician'].map(veri)
job_data['technician'] = job_data['technician'].map({'False': 0, 'True': 1})

job_data['unemployed'] = job_data['unemployed'].map(veri)
job_data['unemployed'] = job_data['unemployed'].map({'False': 0, 'True': 1})

job_data['unknown'] = job_data['unknown'].map(veri)
job_data['unknown'] = job_data['unknown'].map({'False': 0, 'True': 1})
```

```python
data.shape
```

```python
job_data.shape
```

```python
features  = pd.concat([data, job_data], axis=1)
```

```python
features.drop(columns=['job','y'], inplace=True)
```

```python
features['default'] = features['default'].map({'no': -1, 'yes': 1, 'unknown': 0})
```

```python
features.isnull().sum()
```

```python
features.columns
```

```python
X = features
Y = data['y']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy= 0.25)  
X_train, y_train = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(C= 12)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1-Score: {f1:.2f}')
```
